```python
import numpy as np
from sklearn.linear_model import LinearRegression
import joblib

# 1. Create some simple data (X is input, y is the number to predict)
# Let's say: if X is 1, y is 3; if X is 2, y is 5... (y = 2x + 1)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 5, 7, 9, 11])

# 2. Initialize and Train the model
niyas_model = LinearRegression()
niyas_model.fit(X, y)

# 3. Test it: Predict the value for 10 (should be 21)
prediction = niyas_model.predict([[10]])
print(f"Prediction for 10: {prediction[0]}")

# 4. NOW save it using the correct variable name
joblib.dump(niyas_model, 'model.pkl')
print("Model saved successfully as model.pkl!")
```

```python
import streamlit as st
import joblib
import numpy as np

# 1. Set the Title of your website
st.title("Niyas's Number Predictor")
st.write("This app uses a Linear Regression model to predict numbers!")

# 2. Load the model you saved earlier
model = joblib.load('model.pkl')

# 3. Create a user input box
user_input = st.number_input("Enter a number to predict:", value=0.0)

# 4. Make the prediction when the button is clicked
if st.button("Predict"):
    # Reshape the input to match the model's expected format
    prediction = model.predict(np.array([[user_input]]))
    st.success(f"The predicted result is: {prediction[0]:.2f}")
    
```
